# Facebook Auth

```python
#Program Imports
from selenium import webdriver
from getpass import getpass

#Initialize chrome driver in Incognito Mode
options = webdriver.ChromeOptions()
options.add_argument('--incognito')
driver = webdriver.Chrome(executable_path='./chromedriver/chromedriver.exe', options=options)

#Open the facebook webpage
driver.get("https://www.facebook.com/")
```

```python
#Identify needed elements
username_textbox = driver.find_element_by_id('email')
password_textbox = driver.find_element_by_id('pass')
submit_button = driver.find_element_by_name('login')
```

```python
#Get credentials
username = input('Enter your username or email: ')
password = getpass('Enter yout password: ')
```

```python
#Set credentials
username_textbox.send_keys(username)
password_textbox.send_keys(password)
```

```python
#Click submit button
submit_button.click()
```
